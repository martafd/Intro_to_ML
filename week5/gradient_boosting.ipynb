```python
import pandas as pd

df = pd.read_csv('gbm-data.csv')
df.head()
```

```python
X = df.drop(['Activity'], axis=1).as_matrix()
y = df['Activity'].as_matrix()
```

```python
from sklearn.cross_validation import train_test_split
#!/usr/bin/env python -W ignore::DeprecationWarning

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.8, random_state = 241)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
import numpy as np
from sklearn.metrics import log_loss

def sigmoid(y_pred):
    return 1. / (1 + np.exp(-y_pred))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.metrics import log_loss

lr = [1, 0.5, 0.3, 0.2, 0.1]

for i in lr:
    clf = GradientBoostingClassifier(learning_rate=1, n_estimators=250, verbose=True, random_state=241)
    clf.fit(X_train, y_train)

    train_res = []
    for pred in clf.staged_decision_function(X_train):
        train_res.append(log_loss(y_train,[sigmoid(y_pred_train) for y_pred_train in pred]))

    test_res = []
    for j in clf.staged_decision_function(X_test):
        test_res.append(log_loss(y_test, [sigmoid(y_pred_test) for y_pred_test in j]))

    plt.figure()
    plt.plot(test_res, 'r', linewidth=2)
    plt.plot(train_res, 'g', linewidth=2)
    plt.legend(['test', 'train'])
    plt.show()
```

```python
with open('q1.txt', 'w') as f:
    f.write('overfitting')
f.close()
```

```python
clf = GradientBoostingClassifier(learning_rate=0.2, n_estimators=250, verbose=True, random_state=241)
clf.fit(X_train, y_train)

test_res = []
for j in clf.staged_decision_function(X_test):
    test_res.append(log_loss(y_test, [sigmoid(y_pred_test) for y_pred_test in j]))
```

```python
print 'min value', np.min(test_res)
print 'iteration number', np.argmin(test_res)
```

```python
with open('q2.txt', 'w') as f:
    f.write(str(np.min(test_res)) + ' ' + str(np.argmin(test_res)))
f.close()
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=np.argmin(test_res), random_state=241)
clf.fit(X_train, y_train)
y_pred = clf.predict_proba(X_test)
test_loss = log_loss(y_test, y_pred)
```

```python
with open('q3.txt', 'w') as f:
    f.write(str(test_loss))
f.close()
```
